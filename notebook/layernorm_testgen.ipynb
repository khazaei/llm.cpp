```python
import torch
```

```python
import torch.nn as nn
```

```python
torch.manual_seed(0)
torch.set_printoptions(precision=7)

# NLP Example
batch, sentence_length, embedding_dim = 3, 5, 10
embedding = torch.randn(batch, sentence_length, embedding_dim)
layer_norm = nn.LayerNorm(embedding_dim)
layer_norm.bias = nn.Parameter(torch.randn(embedding_dim))
layer_norm.weight = nn.Parameter(torch.randn(embedding_dim))
# Activate module
out = layer_norm(embedding)
```
